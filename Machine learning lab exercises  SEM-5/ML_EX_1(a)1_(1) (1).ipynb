```python
# 8. Display the outlier values using Z-score (Python code)
import pandas as pd
import numpy as np

# Create a sample dataset
data = {
    "Name": ["John", "Emma", "Michael", "Sophia", "William", "Olivia"],
    "Age": [25, 32, 40, 28, 35, 50],
    "Gender": ["Male", "Female", "Male", "Female", "Male", "Female"],
    "Billing Amount": [20000, 30000, 40000, 50000, 60000, 70000],
    "Room Number": [101, 202, 303, 404, 505, 606]
}

df = pd.DataFrame(data)

# Add the dataset to the existing healthcare dataset
df = pd.concat([df, pd.read_csv("healthcare_dataset.csv")])

#Z-scores
z_scores = np.abs((df.select_dtypes(include=np.number)) - df.select_dtypes(include=np.number).mean()) / df.select_dtypes(include=np.number).std()

z_score_threshold = 3

outliers = df[(z_scores >= z_score_threshold).any(axis=1)]

# Interquartile Range (IQR)
Q1 = df.select_dtypes(include=np.number).quantile(0.25)
Q3 = df.select_dtypes(include=np.number).quantile(0.75)
IQR = Q3 - Q1

# Display the outlier values
print("Outlier values:")
print(outliers)

# Display the IQR for each numeric column
print("Interquartile Range (IQR):")
print(IQR)
```

```python
# 9. Display the features with high positive correlation, high negative correlation, and no correlation
corr_matrix = df.select_dtypes(include=np.number).corr()

high_positive_corr = corr_matrix[corr_matrix > 0.8]
high_negative_corr = corr_matrix[corr_matrix < -0.8]
no_corr = corr_matrix[(corr_matrix >= -0.2) & (corr_matrix <= 0.2)]

print("high positive correlation:")
print(high_positive_corr)

print("high negative correlation:")
print(high_negative_corr)

print("no correlation:")
print(no_corr)
```

```python
# 10. Analyze the skewness of the features using plot distribution graph and display features with right skew, left skew, and no skew
skewness = df.select_dtypes(include=np.number).skew()
right_skew = skewness[skewness > 1]
left_skew = skewness[skewness < -1]
no_skew = skewness[(skewness >= -1) & (skewness <= 1)]

for feature in right_skew.index:
    sns.displot(df[feature], kde=True)
    plt.title(f"Right Skew - {feature}")
    plt.show()

for feature in left_skew.index:
    sns.displot(df[feature], kde=True)
    plt.title(f"Left Skew - {feature}")
    plt.show()

for feature in no_skew.index:
    sns.displot(df[feature], kde=True)
    plt.title(f"No Skew - {feature}")
    plt.show()
```

```python
# 10. Analyze the skewness of the features using plot distribution graph and display features with right skew, left skew, and no skew
skewness = df.select_dtypes(include=np.number).skew()
right_skew = skewness[skewness > 1]
left_skew = skewness[skewness < -1]
no_skew = skewness[(skewness >= -1) & (skewness <= 1)]

for feature in right_skew.index:
    plt.hist(df[feature], bins=30 ,edgecolor="black")
    plt.title(f"Right Skew - {feature}")
    plt.show()

for feature in left_skew.index:
    plt.hist(df[feature], bins=30 ,edgecolor="black")
    plt.title(f"Left Skew - {feature}")
    plt.show()

for feature in no_skew.index:
    plt.hist(df[feature], bins=30,edgecolor="black")
    plt.title(f"No Skew - {feature}")
    plt.show()
```

```python
# 14. Perform bivariate analysis between continuous variable and categorical variable using categorical plot
fig, ax = plt.subplots()
boxplot_data = []
labels = []
for condition in df["Medical Condition"].unique():
    boxplot_data.append(df[df["Medical Condition"] == condition]["Billing Amount"])
    labels.append(condition)

ax.boxplot(boxplot_data, labels=labels)
plt.xticks(rotation=90)
plt.title("Bivariate Analysis - Medical Condition vs. Billing Amount")
plt.show()
```
