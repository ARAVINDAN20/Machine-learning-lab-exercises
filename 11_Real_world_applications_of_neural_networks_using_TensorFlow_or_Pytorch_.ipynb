# Experiment: Real-World Applications of Neural Networks Using TensorFlow

## Aim
To implement and explore a real-world application of neural networks using TensorFlow, focusing on image classification.

## Objectives
- Understand the architecture and implementation of neural networks for solving real-world problems.
- Apply TensorFlow to classify images from the CIFAR-10 dataset.
- Evaluate and visualize the performance of the trained model.

## Tools Used
- **TensorFlow**: For building and training the neural network.
- **Matplotlib** and **Seaborn**: For visualizations.
- **NumPy**: For numerical operations.

## Implementation

### Step 1: Import Libraries
```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Load and Preprocess Data
```python
# Load the CIFAR-10 dataset
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# Normalize the pixel values
X_train = X_train / 255.0
X_test = X_test / 255.0

# Convert labels to one-hot encoding
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Display sample images
class_names = ["Airplane", "Automobile", "Bird", "Cat", "Deer", "Dog", "Frog", "Horse", "Ship", "Truck"]
plt.figure(figsize=(10, 5))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(X_train[i])
    plt.title(class_names[np.argmax(y_train[i])])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

### Step 3: Build the Neural Network Model
```python
# Initialize the model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Display the model architecture
model.summary()
```

### Step 4: Train the Model
```python
# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=64, validation_data=(X_test, y_test))
```

### Step 5: Evaluate the Model
```python
# Evaluate on the test set
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy:.4f}")

# Plot training history
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy Over Epochs')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Over Epochs')
plt.legend()

plt.tight_layout()
plt.show()
```

### Step 6: Visualize Predictions
```python
# Predict on test data
predictions = model.predict(X_test)

# Visualize some predictions
plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.imshow(X_test[i])
    plt.title(f"True: {class_names[np.argmax(y_test[i])]}\nPredicted: {class_names[np.argmax(predictions[i])]}\n")
    plt.axis('off')
plt.tight_layout()
plt.show()
```

### Step 7: Save the Model
```python
# Save the trained model
model.save('cifar10_model.h5')
print("Model saved successfully!")
```

### Step 8: Summary and Observations
```python
print("\nSummary:")
print("1. A convolutional neural network was implemented for the CIFAR-10 image classification task.")
print("2. The model achieved a test accuracy of {:.2f}%, indicating effective learning.".format(test_accuracy * 100))
print("3. Training history and predictions were visualized to analyze model performance.")
